Explode movie_df to avoid the problem of sparcity.

```python
movie_df['genres'] = movie_df['genres'].str.split('|')
ex_movie_df = movie_df.explode('genres')

print('Total no. of genres:', ex_movie_df.genres.nunique())
ex_movie_df.genres.value_counts()
```

```python
ex_movie_df.head(30)
```

Merge exploded movie df to ratings df for analysis of ratings by genre.

```python
merged_ratings_df = pd.merge(ratings_df, ex_movie_df, on = 'movieId', how = 'left')

print(merged_ratings_df.shape)
merged_ratings_df.head(20)
```

```python
from surprise import Reader, Dataset, SVD, accuracy

# Data preparation
reader = Reader(rating_scale = (0.5,5.0))
data = Dataset.load_from_df(merged_ratings_df[['userId', 'movieId', 'rating']], reader)
```

Train test split

```python
from surprise.model_selection import train_test_split

# Split the data into trainvalidation and test sets.
trainset, testset = train_test_split(data, test_size = 0.2, random_state = 42)
```

Model training

```python
algo = SVD()
algo.fit(trainset)
```

Model evaluation

```python
from surprise import accuracy

# Use val test set
predictions = algo.test(testset)
accuracy.rmse(predictions)
```

Predictions

```python
predictions_df = pd.DataFrame(predictions, columns= ['user_id', 'item_id', 'true_rating', 'estimated_rating', 'details'])

predictions_df.head(10)
```

```python
# user and movie ids from merged_ratings_df
user_ids = merged_ratings_df['userId'].nunique()
movie_ids = merged_ratings_df['movieId'].nunique()

print(user_ids)
print(movie_ids)
```

predictons for unrated movies for each user.

```python
user_ids = merged_ratings_df['userId'].unique()
movie_ids = merged_ratings_df['movieId'].unique()

# Number of reccomendations to make
N = 5
recommendations = {}

# Loop over each user
for user_id in user_ids:
    rated_movie_ids = merged_ratings_df[merged_ratings_df['userId'] == user_id]['movieId'].unique()

    unrated_movie_ids = [m_id for m_id in movie_ids if m_id not in rated_movie_ids]

    # Predict ratings for the unrated movies
    predictions = [algo.predict(user_id, m_id) for m_id in unrated_movie_ids]

    # Sort predictions by estimated rating (highest first)
    predictions.sort(key=lambda x: x.est, reverse=True)

    top_n_recommendations = predictions[:N]

    # Store the recommendations for each user
    recommendations[user_id] = [(pred.iid, pred.est) for pred in top_n_recommendations]
```

```python
# Display top 5 recommendations for all users
for user_id, recs in recommendations.items():
    print(f'\nTop {N} recommendations for User {user_id}:')
    for movie_id, predicted_rating in recs:
        print(f'Movie ID: {movie_id}, Predicted Rating: {predicted_rating:.2f}')
```
