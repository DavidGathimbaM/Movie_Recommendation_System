**3. Recommendation Systems Algorithm.**

- KNN algorithm

*a. User based collaborative filtering*

```python
# Create User-Item Matrix
user_movie_matrix = merged_ratings_df.pivot_table(index='userId', columns='movieId', values='rating').fillna(0)
user_movie_matrix
```

Train-test split

```python
from sklearn.model_selection import train_test_split

trainset, testset = train_test_split(merged_ratings_df, test_size= 0.2, random_state= 42)

# Train test matrices from the split data
train_matrix = trainset.pivot_table(index= 'userId', columns= 'movieId', values= 'rating').fillna(0)
test_matrix = trainset.pivot_table(index= 'userId', columns= 'movieId', values= 'rating').fillna(0)
```

Model selection

```python
from sklearn.neighbors import NearestNeighbors

knn = NearestNeighbors(metric = 'cosine', algorithm= 'brute', n_neighbors= 6)
knn.fit(train_matrix)
```

```python
# K-nearest neighbors for users in test_matrix
distances, indices = knn.kneighbors(test_matrix, n_neighbors= 6)

# Loop through each user index to print the organized output
for user_index in range(len(distances)):
    print(f"User ID {user_index + 1}:")
    
    for neighbor_index, distance in zip(indices[user_index][1:], distances[user_index][1:]):
        print(f"  Similar Users: {neighbor_index}, Similarity distance: {distance:.4f}")

    print()
```

predict ratings

```python
import numpy as np

def predict_ratings(user_index, train_matrix, distances, indices):
    similar_users = indices[user_index][1:]
    similar_distances = distances[user_index][1:]
    
    # Weighted ratings
    weighted_sum = np.zeros(train_matrix.shape[1])
    sum_of_weights = np.zeros(train_matrix.shape[1])
    
    for neighbor_index, distance in zip(similar_users, similar_distances):
        if distance > 0:
            weighted_sum += (train_matrix.iloc[neighbor_index] / distance)
            sum_of_weights += (1 / distance)
            
    # Predicted ratings
    predicted_ratings = weighted_sum / sum_of_weights
    return predicted_ratings


predictions = []

# Loop through each user in the test matrix
for user_index in range(len(test_matrix)):
    predicted_ratings = predict_ratings(user_index, train_matrix, distances, indices)
    predictions.append(predicted_ratings)

predictions_df = pd.DataFrame(predictions, columns=train_matrix.columns)
```

compare predictions to actual ratings.

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Extract actual ratings from the test matrix and convert to numpy array
actual_ratings = test_matrix.values
predictions_array = predictions_df.values

# Compare only where ratings exist in the test set
mask = actual_ratings != 0  
actual_ratings = actual_ratings[mask]
predictions_array = predictions_array[mask]


mae = mean_absolute_error(actual_ratings, predictions_array)
print(f"Mean Absolute Error: {mae:.4f}")

rmse = np.sqrt(mean_squared_error(actual_ratings, predictions_array))
print(f"Root Mean Squared Error: {rmse:.4f}")
```

- Hybrid Model (*User and Item based combined*)

Utilised knn in surprise.

```python
from surprise import Dataset, Reader

# Define a Reader object
reader = Reader(rating_scale=(1, 5))

# Load the dataset into Surprise
dataset = Dataset.load_from_df(merged_ratings_df[['userId', 'movieId', 'rating']], reader)
```

```python
from surprise.model_selection import train_test_split

# Split the dataset into train and test sets
trainset, testset = train_test_split(dataset, test_size=0.2, random_state=42)
```

```python
from surprise import KNNBasic

# User-based collaborative filtering
user_based_algo = KNNBasic(sim_options={'user_based': True})
user_based_algo.fit(trainset)
```

```python
# Item-based collaborative filtering
movie_based_algo = KNNBasic(sim_options={'user_based': False})
movie_based_algo.fit(trainset)
```

```python
# Predict ratings using both algorithms
user_based_predictions = user_based_algo.test(testset)
movie_based_predictions = movie_based_algo.test(testset)

# Attributes for user-based predictions
user_based_predictions_df = pd.DataFrame({
    'userId': [pred.uid for pred in user_based_predictions],
    'movieId': [pred.iid for pred in user_based_predictions],
    'user_based_prediction': [pred.est for pred in user_based_predictions]
})

# Attributes for item-based predictions
movie_based_predictions_df = pd.DataFrame({
    'userId': [pred.uid for pred in movie_based_predictions],
    'movieId': [pred.iid for pred in movie_based_predictions],
    'movie_based_prediction': [pred.est for pred in movie_based_predictions]
})

merged_predictions = pd.merge(user_based_predictions_df, movie_based_predictions_df, on=['userId', 'movieId'])
```

```python
# Average predictions from both algorithms
merged_predictions['final_prediction'] = (
    merged_predictions['user_based_prediction'] + merged_predictions['movie_based_prediction']
) / 2

merged_predictions[['userId', 'movieId', 'final_prediction']]
```

```python
from surprise import accuracy

# user-based 
user_based_rmse = accuracy.rmse(user_based_predictions)
print(f'User-Based RMSE: {user_based_rmse}')

# item-based
movie_based_rmse = accuracy.rmse(movie_based_predictions)
print(f'Item-Based RMSE: {movie_based_rmse}')
```

**3. Recommendation Systems Algorithm.**

- Model based

*a. SVD using Surprise*

```python
from surprise import Reader, Dataset, SVD, accuracy

# Data preparation
reader = Reader(rating_scale = (0.5,5.0))
data = Dataset.load_from_df(merged_ratings_df[['userId', 'movieId', 'rating']], reader)
```

```python
# Split the data into train and test sets.
trainset, testset = train_test_split(data, test_size = 0.2, random_state = 42)
```

```python
algo = SVD(random_state = 42)
algo.fit(trainset)
```

```python
from surprise import accuracy

# Use val test set
predictions = algo.test(testset)
accuracy.rmse(predictions)
```

```python
predictions_df = pd.DataFrame(predictions, columns= ['user_id', 'item_id', 'true_rating', 'estimated_rating', 'details'])

predictions_df.head(10)
```

```python
user_ids = merged_ratings_df['userId'].unique()
movie_ids = merged_ratings_df['movieId'].unique()

# Number of reccomendations to make
N = 5
recommendations = {}

# Loop over each user
for user_id in user_ids:
    rated_movie_ids = merged_ratings_df[merged_ratings_df['userId'] == user_id]['movieId'].unique()

    unrated_movie_ids = [m_id for m_id in movie_ids if m_id not in rated_movie_ids]

    # Predict ratings for the unrated movies
    predictions = [algo.predict(user_id, m_id) for m_id in unrated_movie_ids]

    # Sort predictions by estimated rating (highest first)
    predictions.sort(key=lambda x: x.est, reverse=True)

    top_n_recommendations = predictions[:N]

    # Store the recommendations for each user
    recommendations[user_id] = [(pred.iid, pred.est) for pred in top_n_recommendations]
```

```python
# Display top 5 recommendations for all users
for user_id, recs in recommendations.items():
    print(f'\nTop {N} recommendations for User {user_id}:')
    for movie_id, predicted_rating in recs:
        print(f'Movie ID: {movie_id}, Predicted Rating: {predicted_rating:.2f}')
```
